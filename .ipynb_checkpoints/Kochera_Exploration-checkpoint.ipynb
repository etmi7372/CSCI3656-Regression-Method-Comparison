```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def testFunc(x):
    return np.sin(x)

X = np.linspace(0, 2*np.pi, 100)
Y = testFunc(X)

noise = np.random.normal(loc=5, scale=0.1, size=Y.shape)
Ynoise = Y + noise

plt.plot(X, Y, label='Original')
plt.plot(X, Ynoise, label='Noisy')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Generate some sample data
np.random.seed(42)
dates = pd.date_range('2022-01-01', periods=365)
data = pd.Series(np.random.randint(10, 20, size=365), index=dates)

# Fit the ExponentialSmoothing model
model = ExponentialSmoothing(data, trend='add', seasonal='add', seasonal_periods=7)
fit = model.fit()

# Make forecasts for the next 30 days
forecast = fit.forecast(30)

# Plot the actual data and the forecasts
plt.plot(data.index, data, label='Actual')
plt.plot(forecast.index, forecast, label='Forecast')
plt.legend()
plt.show()
```
