```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def testFunc(x):
    return np.sin(x)

X = np.linspace(0, 2*np.pi, 100)
Y = testFunc(X)

noise = np.random.normal(loc=5, scale=0.1, size=Y.shape)
Ynoise = Y + noise

plt.plot(X, Y, label='Original')
plt.plot(X, Ynoise, label='Noisy')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the true function to be fitted
def true_function(x):
    return np.sin(2*np.pi*x)

# Generate some noisy data based on the true function
x_data = np.linspace(0, 1, 50)
y_data = true_function(x_data) + 0.2*np.random.randn(50)

# Define the models to be tested
def linear_model(x, a, b):
    return a*x + b

def quadratic_model(x, a, b, c):
    return a*x**2 + b*x + c

def cubic_model(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

def sine_model(x, a, b, c):
    return a*np.sin(b*x + c)

# Fit each model to the data
linear_params, _ = curve_fit(linear_model, x_data, y_data)
quadratic_params, _ = curve_fit(quadratic_model, x_data, y_data)
cubic_params, _ = curve_fit(cubic_model, x_data, y_data)
sine_params, _ = curve_fit(sine_model, x_data, y_data)

# Plot the results
plt.plot(x_data, y_data, 'o', label='Data')
plt.plot(x_data, true_function(x_data), 'k-', label='True Function')
plt.plot(x_data, linear_model(x_data, *linear_params), label='Linear Model')
plt.plot(x_data, quadratic_model(x_data, *quadratic_params), label='Quadratic Model')
plt.plot(x_data, cubic_model(x_data, *cubic_params), label='Cubic Model')
plt.plot(x_data, sine_model(x_data, *sine_params), label='Sine Model')
plt.legend()
plt.show()
```
